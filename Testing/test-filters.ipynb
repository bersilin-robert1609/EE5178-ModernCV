```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def show_image(image):
    plt.imshow(image, cmap='gray')
    plt.show()
```

```python
image = np.random.random((32, 32))
show_image(image)
```

```python
gaussian_image = np.random.normal(loc=0, scale=1, size=(32, 32))
show_image(gaussian_image)
```

```python
gaussian_image_1 = np.random.normal(loc=0, scale=0.05, size=(64, 64))
gaussian_image_2 = np.random.normal(loc=0, scale=0.1, size=(64, 64))
gaussian_image_3 = np.random.normal(loc=0, scale=0.2, size=(64, 64))

gaussian_image_blurred_1 = cv2.GaussianBlur(gaussian_image_1, (3, 3), 1)
gaussian_image_blurred_2 = cv2.GaussianBlur(gaussian_image_2, (3, 3), 1)
gaussian_image_blurred_3 = cv2.GaussianBlur(gaussian_image_3, (3, 3), 1)

gaussian_image_blurred_4 = cv2.GaussianBlur(gaussian_image_1, (3, 3), 2)
gaussian_image_blurred_5 = cv2.GaussianBlur(gaussian_image_2, (3, 3), 2)
gaussian_image_blurred_6 = cv2.GaussianBlur(gaussian_image_3, (3, 3), 2)

fig, axes = plt.subplots(nrows=3, ncols=3)

axes = axes.flatten()

axes[0].imshow(gaussian_image_1, cmap='gray')
axes[1].imshow(gaussian_image_2, cmap='gray')
axes[2].imshow(gaussian_image_3, cmap='gray')

axes[3].imshow(gaussian_image_blurred_1, cmap='gray')
axes[4].imshow(gaussian_image_blurred_2, cmap='gray')
axes[5].imshow(gaussian_image_blurred_3, cmap='gray')

axes[6].imshow(gaussian_image_blurred_4, cmap='gray')
axes[7].imshow(gaussian_image_blurred_5, cmap='gray')
axes[8].imshow(gaussian_image_blurred_6, cmap='gray')

plt.show()
```
